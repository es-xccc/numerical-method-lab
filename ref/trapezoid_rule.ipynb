```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def function(x):
    f = 4*(x**2) + (1/3)*x + 5
    return f
```

```python
def trapezoid_rule(n):
    a = -2
    b = 4
    h = (b - a) / (n - 1)
    x = np.linspace(a,b,n)
    f=function(x)
    I_trap = (h/2)*(f[0] + 2*sum(f[1:n-1]) + f[n-1])
    I_trap = math.floor(I_trap*1000)/1000
    err_trap =128 - I_trap
    x_curve = np.linspace(a, b, 1000)
    plt.plot(x_curve , function(x_curve), linewidth=2, c = 'r')
    for i in range(n-1):
        if(i%3==0):
            plt.fill("j","k",'darkorange',data={"j":[x[i],x[i+1],x[i+1],x[i]],"k":[0,0,function(x[i+1]),function(x[i])]},alpha=0.5)
        elif(i%3==1):
            plt.fill("j","k",'forestgreen',data={"j":[x[i],x[i+1],x[i+1],x[i]],"k":[0,0,function(x[i+1]),function(x[i])]},alpha=0.5)
        else:
            plt.fill("j","k",'royalblue',data={"j":[x[i],x[i+1],x[i+1],x[i]],"k":[0,0,function(x[i+1]),function(x[i])]},alpha=0.5)
    plt.show()
    print("I_trap:",I_trap)
    print("err_trap:",round(err_trap,3))
```

```python
trapezoid_rule(3)
```

```python
trapezoid_rule(11)
```

```python
trapezoid_rule(21)
```
