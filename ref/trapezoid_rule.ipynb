```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def function(x):
    f = 4*(x**2) + (1/3)*x + 5
    return f
```

```python
def trapezoid_rule(n):
    a, b = -2, 4
    h = (b - a) / (n - 1)
    x = np.linspace(a, b, n)
    f = function(x)
    I_trap = (h / 2) * (f[0] + 2 * sum(f[1:n-1]) + f[n-1])
    I_trap = math.floor(I_trap * 1000) / 1000
    err_trap = 128 - I_trap

    plt.figure(figsize=(6, 4))
    x_curve = np.linspace(a, b, 1000)
    plt.plot(x_curve, function(x_curve), linewidth=2, c='r')

    colors = ['darkorange', 'forestgreen', 'royalblue']
    for i in range(n - 1):
        plt.fill("j", "k", colors[i % 3], data={"j": [x[i], x[i+1], x[i+1], x[i]], "k": [0, 0, function(x[i+1]), function(x[i])]}, alpha=0.5)

    plt.show()

    print(f"I_trap: {I_trap}")
    print(f"err_trap: {round(err_trap, 3)}")
```

```python
trapezoid_rule(3)
```

```python
trapezoid_rule(11)
```

```python
trapezoid_rule(21)
```
