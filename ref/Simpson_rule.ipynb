```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def function(x):
    f = np.exp(np.sin(x))
    return f
```

```python
I_simp=np.zeros(5)
err_simp=np.zeros(5)
for i in range(5):
    n=i*2+3
    a = 0
    b = np.pi
    h = (b - a) / (n - 1)
    x = np.linspace(a, b, n)
    f = function(x)
    I_simp[i] = (h/3) * (f[0] + 2*sum(f[2:n-2:2]) + 4*sum(f[1:n-1:2]) + f[n-1])
    err_simp[i] = abs(6.208758035711 - I_simp[i])

# Plot solution
x = np.linspace(3, 11, 5)

plt.figure(figsize = (12, 8))
plt.xlabel("grid points")
plt.ylabel("error")
plt.plot(x, err_simp, label = "Exact solution")
print("minimum error:", min(err_simp))
plt.legend()
plt.show()

```
