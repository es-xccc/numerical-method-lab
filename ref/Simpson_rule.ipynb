```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def function(x):
    f = np.exp(np.sin(x))
    return f
```

```python
def simpson_rule(n, a, b, function):
    h = (b - a) / (n - 1)
    x = np.linspace(a, b, n)
    f = function(x)
    return (h / 3) * (f[0] + 2 * sum(f[2:n-2:2]) + 4 * sum(f[1:n-1:2]) + f[n-1])

exact_value = 6.208758035711
I_simp = np.zeros(5)
err_simp = np.zeros(5)

for i in range(5):
    n = i * 2 + 3
    a = 0
    b = np.pi
    I_simp[i] = simpson_rule(n, a, b, function)
    err_simp[i] = abs(I_simp[i] - exact_value)

# Plot solution
x = np.linspace(3, 11, 5)

plt.figure(figsize=(12, 8))
plt.xlabel("grid points")
plt.ylabel("error")
plt.plot(x, err_simp, label="Exact solution")
print("minimum error:", min(err_simp))
plt.legend()
plt.show()
```
