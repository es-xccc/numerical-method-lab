```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_central_diff(x, h): 
        x = np.array(x)
        central_diff = (x[2:] - x[:-2]) / (2 * h)
        return central_diff
```

```python

h = np.zeros(15)
max_error = np.zeros(15)
h[0]=0.5 #設置初始h
for i in range(15):#iterations=15
    if i>0:
        h[i]=h[i-1]/2 #計算新的h
    x = np.arange(0, 2*np.pi, h[i])#x軸範圍從0到2pi
    y = np.cos(x)#f(x)=cos(x)
    # compute vector central differences
    central_diff = my_central_diff(y,h[i])
    # compute corresponding grid
    x_diff = x[1:-1:]
    # compute exact solution
    exact_solution = -np.sin(x_diff)
    # Compute max error between
    max_error[i] = max(abs(exact_solution - central_diff))
    
# Plot solution
plt.figure(figsize = (6, 4))
plt.loglog(h,max_error,marker="v",linestyle="None")
plt.xlabel("step size")#設置x軸label
plt.ylabel("max error")#設置y軸label
plt.show()
```
