```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def diff_f(t):
    df = math.exp(-2*t)
    return df
```

```python
a = 0
b = 1
x = np.linspace(a, b, 11)
f = np.zeros(11)
f[0]=-0.5
for i in range(10):
    f[i+1]=f[i]+0.1*diff_f(x[i])

plt.figure(figsize = (12, 8))
plt.title("Euler v.s. Exact")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.plot(x, f, label = "Forward Euler Method")

exact_function = np.zeros(11)
for i in range(11):
    exact_function[i]=-0.5*math.exp(-2*x[i])

plt.plot(x, exact_function, label = "Exact",color = "red", linestyle="--")
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
