```python
import numpy as np
import matplotlib.pyplot as plt

def central_diff(x, h): 
    return (x[2:] - x[:-2]) / (2 * h)

h = 0.1
x = np.arange(0, 2 * np.pi, h)
y = np.sin(x)

central_diff_values = central_diff(y, h)
x_diff = x[1:-1]
exact_solution = np.cos(x_diff)

plt.figure(figsize = (12, 8))
plt.plot(x_diff, central_diff_values, label = "Finite difference approximation", linewidth=8)
plt.plot(x_diff, exact_solution, label = "Exact solution", linewidth=4)
plt.legend()
plt.show()

max_error = np.max(np.abs(exact_solution - central_diff_values))
print('The maximum error is', max_error)
```
