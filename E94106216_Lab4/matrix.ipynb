```python
import random
from copy import deepcopy


class Matrix:
    nr = 0
    nc = 0
    mat = []
    def __init__(self, nrows, ncols):
        """Construct a (nrows X ncols) matrix"""
        self.nr = nrows
        self.nc = ncols
        self.mat = [[random.randint(0, 10) for _ in range(ncols)] for _ in range(nrows)]
        
    
    def add(self, m):
        """return a new Matrix object after summation"""
        # Check two matrix are in the same size
        if (self.nr != m.nr) or (self.nc != m.nc):
            return None
        # add self.matrix with m.matrix
        result = deepcopy(self)
        for i in range(result.nr):
            for j in range(result.nc):
                result.mat[i][j] += m.mat[i][j]
        return result


    def sub(self, m):
        """return a new Matrix object after substraction"""
        # Check two matrix are in the same size
        if (self.nr != m.nr) or (self.nc != m.nc):
            return None
        # substraction self.matrix from m.matrix
        result = deepcopy(self)
        for i in range(result.nr):
            for j in range(result.nc):
                result.mat[i][j] -= m.mat[i][j]
        return result


    def mul(self, m):
        """return a new Matrix object after multiplication"""
        # Check if the nrows of m is the same as ncols of self
        if self.nc != m.nr:
            return None
        # Multiply self.matrix and m.matrix
        result = Matrix(self.nr, m.nc)
        for i in range(self.nr):
            for j in range(m.nc):
                result.mat[i][j] = 0
                for k in range(self.nc):
                    result.mat[i][j] += self.mat[i][k] * m.mat[k][j]
        return result


    def transpose(self):
        """return a new Matrix object after transpose"""
        result = Matrix(self.nc, self.nr)
        # Transpose
        for i in range(self.nr):
            for j in range(self.nc):
                result.mat[j][i] = self.mat[i][j]
        return result
    

    def display(self):
        """Display the content in the matrix"""
        for row in self.mat:
            print(' '.join(str(item) for item in row))
```

```python
a_rows = int(input("Enter A matrix's rows:"))
a_cols = int(input("Enter A matrix's cols:"))
print("Matrix A({}, {}):".format(a_rows, a_cols))
A = Matrix(a_rows, a_cols)
A.display()
    
b_rows = int(input("Enter B matrix's rows:"))
b_cols = int(input("Enter B matrix's cols:"))
print("Matrix B({}, {}):".format(b_rows, b_cols))
B = Matrix(b_rows, b_cols)
B.display()
    
print("="*10, 'A + B', "="*10)
result = A.add(B)
if result is not None:
    result.display()
else:
    print('Matrixs\' size should be in the same size')
    
print("="*10, 'A - B', "="*10)
result = A.sub(B)
if result is not None:
    result.display()
else:
    print('Matrixs\' size should be in the same size')

print("="*10, 'A * B', "="*10)
result = A.mul(B)
if result is not None:
    result.display()
else:
    print(f'Cannot multiply two matrixs with sizes ({A.nr}, {A.nc}) and ({B.nr}, {B.nc})')

print("="*5, "the transpose of A*B", "="*5)
if result is not None:
    result = result.transpose()
    result.display()
```
